```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=100),
    'Open': np.random.uniform(100, 200, 100),
    'High': np.random.uniform(100, 200, 100),
    'Low': np.random.uniform(100, 200, 100),
    'Close': np.random.uniform(100, 200, 100),
    'Adj Close': np.random.uniform(100, 200, 100),
    'Volume': np.random.randint(1000, 5000, 100)
})
data['Date'] = pd.to_datetime(data['Date'])  # rani
```

```python
plt.figure(figsize=(10, 5)) #rani
plt.plot(data['Date'], data['Close'], label='Close Price', color='blue')
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5)) #rani
plt.bar(data['Date'].dt.month, data['Volume'], color='skyblue')
plt.title('Monthly Volume Distribution')
plt.xlabel('Month')
plt.ylabel('Volume')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5)) #rani
ax = fig.add_subplot(111, projection='3d')
x = np.arange(len(data))
y = data['High']
z = data['Low']
ax.bar(x, y, zs=z, zdir='y', color='cyan', alpha=0.8)
ax.set_xlabel('Index')
ax.set_ylabel('High')
ax.set_zlabel('Low')
plt.show()
```

```python
plt.hist(data['Close'], bins=15, color='purple', edgecolor='black')
plt.title('Distribution of Close Prices') #rani
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(data['Volume'], data['Close'], color='red', alpha=0.5)
plt.title('Volume vs Close Price')
plt.xlabel('Volume')
plt.ylabel('Close Price') #rani
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))  #rani
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Open'], data['High'], data['Close'], color='green')
ax.set_xlabel('Open')
ax.set_ylabel('High')
ax.set_zlabel('Close')
plt.show()
```

```python
import matplotlib.pyplot as plt  #rani
import numpy as np
days = pd.date_range(start='2023-01-01', periods=100)
data = pd.DataFrame({
    'Day': days,
    'Open': np.random.uniform(100, 150, 100),
    'Close': np.random.uniform(80, 130, 100)
})
plt.stackplot(data['Day'], data['Open'], data['Close'], labels=['Open', 'Close'])
plt.legend(loc='upper left')
plt.title('Cumulative Trends of Open and Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# rani
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Category Proportions')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].plot(data['Day'], data['Open'], color='blue')
axes[0, 0].set_title('Open Prices Over Time')
axes[0, 1].hist(data['Close'], bins=10, color='green')
axes[0, 1].set_title('Distribution of Close Prices')
axes[1, 0].scatter(data['Open'], data['Close'], color='purple')
axes[1, 0].set_title('Open vs Close')
axes[1, 1].boxplot([data['Open'], data['Close']], labels=['Open', 'Close'])
axes[1, 1].set_title('Price Spread')
plt.tight_layout()
plt.show()  #rani
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x, y = np.meshgrid(range(10), range(10))
z = np.sin(x) + np.cos(y)

ax.plot_wireframe(x, y, z, color='black')
ax.set_title('3D Wireframe Plot')
plt.show()  #rani
```

```python
import seaborn as sns #rani
corr = data[['Open', 'Close']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.jointplot(x='Open', y='Close', data=data, kind='scatter', marginal_kws=dict(bins=10, fill=True))
plt.show()
 #rani
```

```python
sns.violinplot(data=[data['Open'], data['Close']], palette='muted')
plt.xticks([0, 1], ['Open', 'Close'])
plt.title('Violin Plot of Open and Close Prices')
plt.show()  #rani
```

```python
sns.boxplot(data=[data['Open'], data['Close']], palette='Set3')
plt.xticks([0, 1], ['Open', 'Close'])
plt.title('Box Plot of Open and Close Prices')
plt.show() #rani
```

```python
from math import pi #rani
categories = ['A', 'B', 'C', 'D']
values = [4, 3, 2, 5]
N = len(categories)
angles = [n / float(N) * 2 * pi for n in range(N)]
values += values[:1]
angles += angles[:1]
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories)
ax.plot(angles, values)
ax.fill(angles, values, 'b', alpha=0.1)
plt.show()
```
